```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def find_radius(area):
    return np.sqrt(area / np.pi)
```

```python
find_radius(159)
```

```python
# change these ...
category = 'other'
family = 2
one_to_one = 16
just_human = 18
total_genes = family + one_to_one + just_human
print(total_genes)
```

```python
# see colors here ... https://matplotlib.org/stable/gallery/color/named_colors.html
# can change these names
family_color = 'yellowgreen'
human_color = 'green'
one_to_one_color = 'darkgreen'
```

```python
center = (0,0)

total_radius = find_radius(total_genes)
family_radius = find_radius(one_to_one + family)
one_to_one_radius = find_radius(one_to_one)

total_circle = plt.Circle(center, total_radius, color=human_color, label = 'total')
family_circle = plt.Circle(center, family_radius, color=family_color, label = 'family')
tardi_circle = plt.Circle(center, one_to_one_radius, color=one_to_one_color, label = 'one to one')

fig, ax = plt.subplots(dpi=300) # note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)
# fig = plt.gcf()
# ax = fig.gca()

ax.add_patch(total_circle)
ax.add_patch(family_circle)
ax.add_patch(tardi_circle)
ax.set_aspect('equal', adjustable='box')
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
plt.title(category, fontsize=24)
plt.axis('off')
plt.legend()
plt.show()
```
