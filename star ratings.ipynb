```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
from matplotlib.markers import MarkerStyle
from matplotlib.transforms import Affine2D

text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontfamily='monospace')
marker_style = dict(linestyle=':', color='0.8', markersize=10,
                    markerfacecolor="tab:blue", markeredgecolor="tab:blue")


def format_axes(ax):
    ax.margins(0.2)
    ax.set_axis_off()
    ax.invert_yaxis()


def split_list(a_list):
    i_half = len(a_list) // 2
    return a_list[:i_half], a_list[i_half:]
```

```python
fig, ax = plt.subplots()
fig.suptitle('Marker fillstyle', fontsize=14)
fig.subplots_adjust(left=0.4)

filled_marker_style = dict(marker='*', linestyle=':', markersize=15,
                           color='darkgrey',
                           markerfacecolor='tab:blue',
                           markerfacecoloralt='lightsteelblue',
                           markeredgecolor='brown')

for y, fill_style in enumerate(Line2D.fillStyles):
    ax.text(-0.5, y, repr(fill_style), **text_style)
    ax.plot([y] * 3, fillstyle=fill_style, **filled_marker_style)
format_axes(ax)
```

```python
def starRating(ax, category, fillstyles):

    filled_marker_style = dict(marker='*', linestyle=':', # for line that connects the markers
                               markersize=18,
                               color='darkgrey', # line color for line that connects the markers
                               markerfacecolor='tab:blue',
                               markerfacecoloralt='lightsteelblue',
                               markeredgecolor='black')
    num_stars = len(fillstyles)
    y=1
    yaxis_buff = 0.1
    ytext_buff = 0.04
    xlims = [1, 4.5]
    fontsize=20
    start_x = 3.78
    x_span = 0.6
    
    for i,x in enumerate(np.linspace(start_x,start_x+x_span,num_stars)):
        ax.plot(x, y, fillstyle = fillstyles[i], **filled_marker_style)
    ax.set_ylim([y-yaxis_buff,y+yaxis_buff])
    ax.text(1,y-ytext_buff,category, fontsize=fontsize)
    ax.set_xlim(xlims)
    ax.set_axis_off()
    return ax
```

```python
# copper glockenspiel
fig, [ax1,ax2,ax3,ax4] = plt.subplots(4,1,figsize=(4,2), dpi=150)
ax1 = starRating(ax1,'construction',['full','full','full'])
ax2 = starRating(ax2,'musical success',['full','full','full'])
ax3 = starRating(ax3,'math required',['full','full','none'])
ax4 = starRating(ax4,'bling needed',['none','none','none'])
plt.show()
```
